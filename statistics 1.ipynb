#1. Explain the different types of data (qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.

Answer:

**Qualitative Data:** Non-numeric data that describes qualities or characteristics.

        Examples: Colors (red, blue), gender (male, female).

**Quantitative Data:** Numeric data representing quantities.

        Examples: Height (170 cm), age (25 years).

Scales of Measurement:

* Nominal: Categorical with no natural order.
      Example: Blood type (A, B, AB, O).

* Ordinal: Categorical with a meaningful order but unequal intervals.

      Example: Pain levels (low, medium, high).

* Interval: Numeric with equal intervals but no true zero.

      Example: Temperature in Celsius.

* Ratio: Numeric with equal intervals and a true zero.

      Example: Weight, height, income.

#2. What are the measures of central tendency, and when should you use each? Discuss the mean, median, and mode with examples and situations where each is appropriate.

Answer:

* **Mean:** The average of data. Best for symmetric distributions.

      Example: Average marks of students.

* **Median**: The middle value. Best for skewed data or data with outliers.

      Example: Median income in a country.

* **Mode**: Most frequent value. Best for categorical or modal data.

      Example: Most common shoe size sold.



# 3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?

Answer:

*Dispersion: Measures how much data varies around the central value.

*Variance: Average of squared deviations from the mean.

*Standard Deviation: Square root of variance, showing spread in original units.

Higher values indicate more spread out data.

#4. What is a box plot, and what can it tell you about the distribution of data?

Answer:

A box plot (or box-and-whisker plot) visually displays the 5-number summary:

Minimum, Q1, Median (Q2), Q3, Maximum

It reveals:

* Central tendency

* Spread (IQR)

* Presence of outliers

* Skewness

# 5. Discuss the role of random sampling in making inferences about populations.
Answer:

Random sampling ensures that every member of the population has an equal chance of being selected, reducing bias. It allows:

* Accurate estimation of population parameters

* Generalization of sample results to the population

* Valid application of statistical tests

#6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?
Answer:

* Skewness measures the asymmetry of a distribution.

* Positive Skew: Tail on right; mean > median

* Negative Skew: Tail on left; mean < median

* Zero Skew: Symmetrical distribution

* Impact: Skewness affects which measure of central tendency to use and interpretation of spread.

# 7. What is the interquartile range (IQR), and how is it used to detect outliers?
Answer:

IQR = Q3 − Q1

Represents the middle 50% of data.

Outlier Detection:

Lower Bound = Q1 − 1.5 × IQR

Upper Bound = Q3 + 1.5 × IQR
Data outside these bounds are considered outliers.


# 8. Discuss the conditions under which the binomial distribution is used.
Answer:

Use binomial distribution when:

* Fixed number of trials (n)

* Each trial has two outcomes (success/failure)

* Constant probability of success (p)

* Independent trials

      Example: Flipping a coin 10 times.

# 9. Explain the properties of the normal distribution and the empirical rule (68-95-99.7 rule).
Answer:

Properties:

* Bell-shaped, symmetric about the mean

* Mean = Median = Mode

* Defined by mean (μ) and standard deviation (σ)

Empirical Rule:

* 68% within ±1σ

* 95% within ±2σ

* 99.7% within ±3σ

# 11. Explain what a random variable is and differentiate between discrete and continuous random variables.
Answer:

A random variable assigns numerical values to outcomes of a random experiment.

Types:

* Discrete: Countable values (e.g., number of students)

* Continuous: Infinite values in a range (e.g., height, weight)
